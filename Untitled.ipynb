```python
#Ubah semua tipe data pada kolom menjadi string untuk menghindari ambiguitas
data.columns = list(map(str, data.columns))

for x in data.columns:
    print (type(x))
```

```python
#karena tahun sudah dikonversi menjadi string, buat variabel untuk mempermudah
#pemanggilan tahun

years = list(map(str, range(1960,2019)))
years
```

```python
#buat kolom baru Total Spend
data['Total Spend'] = data.iloc[:, 2:].sum(axis=1)

data.head()
```

Kolom Total Spend berisi jumlah pengeluaran negara untuk keperluan militer dari tahun 1960-2018

```python
import matplotlib as mpl

mpl.style.use(['ggplot'])
#Top 15 Negara dengan pengeluaran terbanyak
data_countries = data[data['Type']=='Country']
sorted_data = data_countries.sort_values(by='Total Spend', ascending=False)
sorted_data
sorted_data.set_index('Name', inplace=True)
data_top15_countries = sorted_data['Total Spend'].head(15)
# print(data_top15_countries)

#plot the barplot
plt.figure(figsize=(12,12))
data_top15_countries.plot(kind='barh', sort_columns=True, color='steelblue')
plt.xlabel('Total Pengeluaran (USD)')
plt.ylabel('Negara')
plt.title('Total Pengeluaran Militer selama 1960-2018 berdasarkan Negara')

#annotasi label nilai untuk tiap negara
for index, value in enumerate(data_top15_countries):
    label = format(int(value),',')
    #letakkan teks pada akhir bar dengan mengurangi 47000 dari x
    #dan 0.1 dari y
    plt.annotate(label, xy=(value-48000, index-0.1),
                color='black')
plt.show()
```
